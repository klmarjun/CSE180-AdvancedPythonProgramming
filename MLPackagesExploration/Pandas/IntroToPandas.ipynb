```python
#Checking for Null values in each column
new_df2.isnull().sum()
```

```python
new_df2['Marks']
```

```python
#Returns the mean of the Marks column in the DF
new_df2['Marks'].mean()
```

```python
#Filling nan values with the respective column mean values
new_df2['Marks']=new_df2['Marks'].fillna(new_df2['Marks'].mean())
```

```python
#Yay😎 Null values in the column Marks is now filled.
new_df2
```

```python
#Viewing the unique elements in the column Dept
#This is done to categorize data which has the same meaning but different format
set(new_df2['Dept'])
```

```python
#Wrong Format(TypeCasting)
dic = {'Bsc':'Bachelor of Science','Civil':'Civil Engineering','Cse':'Computer Science Engineering','ECE':'Electronic Communication Engineering','EEE':'Electric and Electronic','Mech':'Mechanical Engineering'}
```

```python
#Converting the respective key to values in the DataFrame using the Map Function
new_df2['Dept'] = new_df2['Dept'].map(dic)
```

```python
#Yay😎 Done with Formatting
new_df2
```
